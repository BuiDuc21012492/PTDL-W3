### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def legal_drinker(age):
    return age > 17
```

```python
stud_alcohol['legal_drinker'] = stud_alcohol['age'].apply(legal_drinker)
print("\nStep 10. DataFrame with 'legal_drinker':\n", stud_alcohol.head())
```

### Step 10. Multiply every number of the dataset by 10.
##### I know this makes no sense, don't forget it is just an exercise

```python
def multiply_by_10(x):
    if isinstance(x, (int, float)):  # Check if numeric
        return x * 10
    return x  # Return original value if not numeric
```

```python
for col in stud_alcohol.columns:
    stud_alcohol[col] = stud_alcohol[col].apply(multiply_by_10)
stud_alcohol.head()
```
